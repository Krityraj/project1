```python
#Linkage
X=df.iloc[:,2:4].values
X
X.shape
```

```python
#scatter plot
plt.scatter(X[:,0],X[:,1],color='red')
plt.ylabel('Qual_miles')
plt.xlabel('cc1_miles')
plt.show()
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7)) 
plt.title('Dendrogram')
dend=shc.dendrogram(shc.linkage(X,method='single'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=7,affinity='euclidean',linkage='single')
Y=cluster.fit_predict(X)
Y_new=pd.DataFrame(Y)
Y_new[0].value_counts()
```

```python
#scatterplot
plt.figure(figsize=(10,7))
plt.scatter(X[:,0],X[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
'''AS we can see that single linkage is not giving a better result because one cluster
is getting maximum linkage and other clusters is geeting few linkages which
is not balancing the data.So we have other options for better results'''
```

```python
#b)
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7)) 
plt.title('Dendrogram')
dend=shc.dendrogram(shc.linkage(X,method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=7,affinity='euclidean',linkage='complete')
Y=cluster.fit_predict(X)
Y_new=pd.DataFrame(Y)
Y_new[0].value_counts()
```

```python
#scatterplot
plt.figure(figsize=(10,7))
plt.scatter(X[:,0],X[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
'''AS we can see that there is improvement then single linkage but still
not giving the balance in the dataset.So we will do further analysis to make data appropriate'''
```

```python
#b) DBscan clustering


X1=df.iloc[:,2:4].values
X1
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS_X=SS.fit_transform(X1)
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=1.2,min_samples=800)
dbscan.fit(SS_X)
```

```python
Y=dbscan.labels_
pd.DataFrame(Y).value_counts()
```

```python
#Create a new columns:-
df['cluster']=pd.DataFrame(Y)
df.head()
df.tail()
noise_points=df[df['cluster']==-1]   
noise_points

final_data=df[df['cluster']==0]
final_data

t=df['cluster'].value_counts()
t
```
