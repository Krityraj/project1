```python
#split
X=df.iloc[:,1:14]
X
Y=df['Type']
Y
```

```python
#Standardisation

from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS_X=SS.fit_transform(X)

from sklearn.preprocessing import LabelEncoder
LE= LabelEncoder()
df['Type']=LE.fit_transform(df['Type'])
df['Type']
```

```python
pd.set_option('display.float_format',lambda x:'%.5f'%x)
```

```python
#Load decomposition to do PCA analysis with sklearn
from sklearn.decomposition import PCA
pca=PCA()
PC=pca.fit_transform(SS_X)
pd.DataFrame(PC)
```

```python
df1=pd.DataFrame(pca.explained_variance_ratio_)
df1
```

```python
df1.iloc[0:1,]*100
df1.iloc[1:2]*100
df1.iloc[2:3]*100
df1.iloc[0:2]*100
```

```python
X1=PC[:,0:2]
X1
```

```python
#Model fitting
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(X1,Y)
```

```python
#model prediction
y_pred=logreg.predict(X1)
```

```python
#Metrics
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(Y,y_pred)
cm
ac1=accuracy_score(Y,y_pred)
ac1
```

```python
# we are getting a very good accuracy score of 0.96.

#lets try it with cluster analysis and check the accuracy
 
#cluster analysis
```

```python
X=df.iloc[:,1:3]
X
```

```python
from sklearn.preprocessing import StandardScaler
SS=StandardScaler()
SS_X=SS.fit_transform(X)
```

```python
#Scatter plot
plt.scatter(SS_X[:,0],SS_X[:,1],color='green')
plt.ylabel('Alcohol')
plt.xlabel('Malic')
plt.show()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
dend=shc.dendrogram(shc.linkage(SS_X,method='single'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='single')
Y=cluster.fit_predict(SS_X)
Y_new=pd.DataFrame(Y)
Y_new[0].value_counts()
```

```python
#scatterplot
plt.figure(figsize=(10,7))
plt.scatter(SS_X[:,0],SS_X[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
#Single linkage is not giving the proper result as one data contains maximum values and others contains very few values
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
dend=shc.dendrogram(shc.linkage(SS_X,method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
Y=cluster.fit_predict(SS_X)
Y_new=pd.DataFrame(Y)
Y_new[0].value_counts()
```

```python
#scatterplot
plt.figure(figsize=(10,7))
plt.scatter(SS_X[:,0],SS_X[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
#so from complete linkage we are getting the good result as the data is divided in a balancing way
#still for more results to see we can go with DBscan
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.75,min_samples=45)
dbscan.fit(SS_X)
```

```python
Y=dbscan.labels_
pd.DataFrame(Y).value_counts()
```
