```python
#convert (string of list) into list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
    
    
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#cast
def convert(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
        
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
#dropping
movies.drop('crew',axis=1,inplace=True)
```

```python
movies.info()
```

```python
movies.head()
```

```python
#since (column='overview')is a string and we will have to convert it in list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#replace gaps between two words
movies['genre']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
#generate new column
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']
```

```python
movies.head()
```

```python
#new column
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
#convert in string
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#Convert in lower
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return" ".join(y)
    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#Vectorisation
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
#here instead of euclidean we will calculate the cosine distance between the vectors to find
#the similarity between the movies
#the more the angle less will be the similarity between the movies
#the less the angle between the two vectors more the similarities
#the higher the dimension means if we are having 5000 or 10000 or more values in the dataset here euclidean distance will fail
#so for that instead of euclidean distance we will use Cosine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)#indicating the similarity of one movie with the rest of all the movies
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
similarity
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
#enumerate
#we use enumerate so that while sorting the vectors we dont loss the realtion with one vectors with another
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
#top 5 movies
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(i[0])
```

```python
recommend('Avatar')
```
